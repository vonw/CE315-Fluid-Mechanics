### Section 10.4 - Turbulent Flow in Pipes

### The following code creates the Resistance coefficient (f) versus Reynolds Number plot (Figure 10.8) for various values of the Relative Roughness, ks/D.  It shows that when the value of ks/D is small, f goes down as Re increases.  This appears to indicate that the flow resistance actually decreases as the flow becomes more turbulent.  As ks/D gets large (and approaches 0.1), there is little change in the flow resistance with increasing turbulence.

```python
# Python code to calculate equation 10.26.
Re_array  = array([3e3, 4e3, 6e3, 8e3, 1e4, 3e4, 4e4, 6e4, 8e4, 1e5, 3e5, 4e5, 6e5, 8e5, 1e6, 3e6, 4e6, 6e6, 8e6, 1e7, 3e7, 4e7, 6e7, 8e7, 1e9])
ksD_array = array([0.00001, 0.00005, 0.0001, 0.0002, 0.0004, 0.0006, 0.0008, 0.001, 0.002, 0.004, 0.006, 0.008, 0.01, 0.015, 0.02, 0.03, 0.04, 0.05])

f   = empty(ksD_array.size*Re_array.size).reshape(ksD_array.size,Re_array.size)
for i,ksD in enumerate(ksD_array):
    for j,Re in enumerate(Re_array):
        f[i,j]   = 0.25 / ( log10( (ksD/3.7) + (5.74/Re**0.9) ) )**2
```

```python
# Python code to re-create Figure 10.8.
figure()
for i in range(ksD_array.size):
    loglog(Re_array,f[i,:])
axis([800, 1e8, 0.007, 0.1])
xlabel('Reynolds number')
ylabel('Resistance coefficient, f')

for k,ksD in enumerate(ksD_array):
    text(2e8,f[k,Re_array.size-1],str(ksD))
```
